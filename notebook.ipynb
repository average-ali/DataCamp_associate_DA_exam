## How does the average closing time for claims differ across locations?
To answer this question we need to find the average 'Time to Close' for each 'Location'. Claims in SAO LUIS take the most time to close on average- more than 1000 days, around double the average time to close claims in NATAL. RECIFE comes second on the list, followed by FORTALEZA.

![avg](avg.jpeg)

The visuals show that there is a noticeable difference in the time it takes to close claims across locations, especially between SAO LUIS and other locations. Among the four locations, FORTALEZA and RECIFE are closest to each other, in terms of the time it takes to close claims. Nonetheless, the difference is still significant. In NATAL, fewest number of claims were registered and the average time to close down claims was the lowest.
